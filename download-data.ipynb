## Yahoo Finance
It seems like Yahoo Finance does not provide a official API anymore to download data, but several libraries and third party APIs exist which different techniques, f.e. web scraping, to download data from finance.yahoo.com

```python
# Download stock symbols
symbols = pdr.get_nasdaq_symbols()
print(f"We have downloaded the symbols for {symbols.shape[0]} stocks on Nasdaq")
symbols.head()
```

```python
# Define a date range
start = dt.datetime(2018, 1, 29)
end = dt.datetime.today()
ticker = 'GOOG'

# Get stock data
df = pdr.data.DataReader(ticker, 'yahoo', start, end)
df.head()
```

```python
# Get stock actions
actions = pdr.data.DataReader(ticker, 'yahoo-actions', start, end)
actions.head()
```

```python
# Get stock dividends
dividends = pdr.data.DataReader(ticker, 'yahoo-dividends', start, end)
dividends.head()
```

```python
df.to_csv("./data/GOOG.csv")
```
